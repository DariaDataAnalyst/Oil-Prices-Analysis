### Importing libraries and data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.cluster import KMeans # Here is where you import the k-means algorithm from scikit-learn.
import pylab as pl # PyLab is a convenience module that bulk imports matplotlib.
```

```python
# This option ensures the graphs you create are displayed in your notebook without the need to "call" them specifically.

%matplotlib inline
```

```python
# Import data set
path=r'/Users/dariaperestiuk/Desktop/HW Data Analyst/Oil Price Project'
df=pd.read_csv(os.path.join(path, '01 Data', 'Prepared Data', 'cleaned_merged_with_oilprice.csv'))
```

```python
df.shape
```

```python
df.head()
```

```python
#Create a list of necessary columns
columns = [' Crude oil production', 'Oil demand', 'Crude oil exports',
       'Imports of crude oil', 'Oil Price']
```

```python
#Create a subset
df_filtered = df[columns]
```

```python
# Check for missing values

df_filtered.isnull().sum()
```

```python
# Impute missing values
df_filtered[' Crude oil production'] = df_filtered[' Crude oil production'].fillna(0)
```

```python
df_filtered['Imports of crude oil'] = df_filtered['Imports of crude oil'].fillna(0)
df_filtered['Oil demand'] = df_filtered['Oil demand'].fillna(0)
df_filtered['Crude oil exports'] = df_filtered['Crude oil exports'].fillna(0)
```

### 2. The elbow technique

```python
num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
score = [kmeans[i].fit(df_filtered).score(df_filtered) for i in range(len(kmeans))] # Creates a score that represents 
# a rate of variation for the given cluster option.

score
```

```python
# Plot the elbow curve using PyLab.

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

#### There's a large jump from two to five on the x-axis, but after that, the curve straightens out. This means that the optimal count for our clusters is five.

### 3. k-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 5) 
```

```python
# Fit the k-means object to the data.

kmeans.fit(df_filtered)
```

```python
df_filtered['Clusters'] = kmeans.fit_predict(df_filtered)
```

```python
df_filtered.head()
```

```python
df_filtered['Clusters'].value_counts()
```

```python
# Plot the clusters for the "Crude oil production" and "Oil Price" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_filtered[' Crude oil production'], y=df_filtered['Oil Price'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Oil production') # Label x-axis.
plt.ylabel('Oil Price') # Label y-axis.
plt.show()
```

#### Cluster 0 appears to be in the lower Oil Production range from 0 to 1000, cluster 1 in the mid-range from 1000 to about 4000. Both clusters represent the main density of Oil Production. Cluster 2, 3, 4 represent the high Oil Production. 

```python
# Plot the clusters for the "Oil demand" and "Oil Price" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_filtered['Oil demand'], y=df_filtered['Oil Price'], hue=kmeans.labels_, s=100)

ax.grid(False) 
plt.xlabel('Oil demand') 
plt.ylabel('Oil Price') 
plt.show()
```

#### The majority of data points are concentrated in the 0 to 3000 Oil Demand range despite the Oil Price. However, we can see another major density of 5000 Oil Demand for the lowest Oil Prices

```python
df_filtered.loc[df_filtered['Clusters'] == 4, 'Cluster'] = 'cluster4'
df_filtered.loc[df_filtered['Clusters'] == 3, 'Cluster'] = 'cluster3'
df_filtered.loc[df_filtered['Clusters'] == 2, 'Cluster'] = 'cluster2'
df_filtered.loc[df_filtered['Clusters'] == 1, 'Cluster'] = 'cluster2'
df_filtered.loc[df_filtered['Clusters'] == 0, 'Cluster'] = 'cluster0'
```
